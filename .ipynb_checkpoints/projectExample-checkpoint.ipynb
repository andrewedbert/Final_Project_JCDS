```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import folium as fol
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df['last_review'] = pd.to_datetime(df['last_review'])
df['last_review'] = df['last_review'].fillna(method='ffill')
# df.head()
df['reviews_per_month'] = df['reviews_per_month'].fillna(
    df['reviews_per_month'].mean()
)

df['name'] = df['name'].fillna(df['name'].mode()[0])
df['host_name'] = df['host_name'].fillna(df['host_name'].mode()[0])

df.head()
```

```python
dfBrooklyn = df[df['neighbourhood_group'] == 'Brooklyn']
dfManhattan = df[df['neighbourhood_group'] == 'Manhattan']
dfQueens = df[df['neighbourhood_group'] == 'Queens']
dfStaten = df[df['neighbourhood_group'] == 'Staten Island']
dfBronx = df[df['neighbourhood_group'] == 'Bronx']
```

```python
map = fol.Map(
    location = [40.7, -74.0],
    zoomstart = 10
)

for i in dfBrooklyn.values.tolist()[:1000]:
    fol.Marker(
        location = (i[6], i[7]),
        tooltip = i[1],
        popup = i[1],
        icon = fol.Icon(color = 'red')
    ).add_to(map)
    
for i in dfManhattan.values.tolist()[:1000]:
    fol.Marker(
        location = (i[6], i[7]),
        tooltip = i[1],
        popup = i[1],
        icon = fol.Icon(color = 'blue')
    ).add_to(map)
    
for i in dfQueens.values.tolist()[:1000]:
    fol.Marker(
        location = (i[6], i[7]),
        tooltip = i[1],
        popup = i[1],
        icon = fol.Icon(color = 'green')

    ).add_to(map)
    
for i in dfStaten.values.tolist()[:1000]:
    fol.Marker(
        location = (i[6], i[7]),
        tooltip = i[1],
        popup = i[1],
        icon = fol.Icon(color = 'orange')
    ).add_to(map)
    
for i in dfBronx.values.tolist()[:1000]:
    fol.Marker(
        location = (i[6], i[7]),
        tooltip = i[1],
        popup = i[1],
        icon = fol.Icon(color = 'pink')
    ).add_to(map)

map.save('airmap.html')
```

```python
from sklearn.preprocessing import StandardScaler
col = ['price', 'minimum_nights', 'calculated_host_listings_count', 'availability_365']
ss = StandardScaler()
df2[col] = ss.fit_transform(df2[col])

# df2 = df2.drop(['id', 'host_id', 'latitude', 'longitude',
#                  'number_of_reviews', 'reviews_per_month'],
#                axis=1)
# df2 = df2.drop(['name', 'host_name'], axis=1)
df2.columns
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

```python
xtr, xtr, yts, ytr = train_test_split(
    df2.drop(columns=['last_review', 'name', 'host_name', 'price', 'neighbourhood', 'id', 'host_id', 'latitude', 'longitude', 'number_of_reviews', 'reviews_per_month']),
    df2['price'],
    test_size=.1
)
```

```python
print(
    np.mean(cross_val_score(LinearRegression(), xtr, ytr, cv=10)) * 100
)
```

```python
print(
    np.mean(cross_val_score(Ridge(alpha=1e-15, normalize=True, max_iter=10000000), xtr, ytr, cv=10)) * 100
)
```

```python
# create model
model = LinearRegression()
model.fit(xtr, ytr)

import pickle
with open('airmodel.pkl', 'wb') as mymodel:
    pickle.dump(model, mymodel)
```
